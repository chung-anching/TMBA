```python
import pandas as pd
```

```python
data = pd.read_csv("上下市櫃普通股_0050_0051.txt", encoding = 'ANSI', low_memory=False)
data["年月日"] = pd.to_datetime(data["年月日"], format = '%Y%m%d')
```

```python
data.head()
```

```python
data_pivot = data.pivot(index = "年月日", columns = "證券代碼") #把dataframe的欄位轉成index是時間，欄是股票代碼
```

```python
data_pivot.head()
```

```python
col = list(data_pivot.columns.levels[0])
```

```python
print(col)
```

```python
for c in col:
    print(c)
    data_pivot[c].to_csv("TEJ_調整後股價資料\\" + c + ".csv", encoding = "utf-8")  #把col裡有的欄位，拆分成好幾個excel檔
```
